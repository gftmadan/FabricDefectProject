```python
DF = pd.DataFrame(columns = ["Image_Name","Image_Width","Image_Height","Image_Location"])
```

```python
def read_the_dataset(dataset_loc):

    for file in os.listdir(dataset_loc):

        image_name = file.split(".")[0]
        
        image_loc = os.path.join(dataset_loc,file)
        img = cv2.imread(image_loc)

        h,w,_ = img.shape

        DF.loc[len(DF)] = {"Image_Name":image_name,"Image_Location":image_loc,"Image_Width":w,"Image_Height":h}
```

```python
read_the_dataset(HOLE_DATASET_LOC)
```

```python
def process_and_store_images(project_name):

    temp_folder = r"D:\CountAI_Projects\Datasets\archive\Exaperimental_Data"

    output_loc = os.path.join(temp_folder,project_name)

    if not os.path.exists(output_loc):
        os.mkdir(output_loc)

    for iter,row in df.iterrows():

        image_loc = row["Image_Location"]
        out_img_loc = os.path.join(output_loc,f"{iter+1}.png")
        shutil.copy2(image_loc,out_img_loc)
```

```python
process_and_store_images("Expriment1")
```
